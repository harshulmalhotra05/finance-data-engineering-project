```python
raw_path = "abfss://finance-data-container@financialpipelineadls2.dfs.core.windows.net/finance-raw/"
```

```python
display(dbutils.fs.ls(raw_path))
```

```python
df_acct = spark.read.csv(raw_path + "finance_account_data_1.csv",
                         header=True,
                         inferSchema=True)

df_cust = spark.read.csv(raw_path + "finance_customer_data_1.csv",
                         header=True,
                         inferSchema=True)

df_loan = spark.read.csv(raw_path + "finance_loan_data_1.csv",
                         header=True,
                         inferSchema=True)

df_tran = spark.read.csv(raw_path + "finance_transaction_data_1.csv",
                         header=True,
                         inferSchema=True)

df_ledg = spark.read.csv(raw_path + "finance_ledger_data_1.csv",
                         header=True,
                         inferSchema=True)
print("All files read Properly")
```

```python
display(df_acct)
```

```python
# Bronze folder
bronze_path = "abfss://finance-data-container@financialpipelineadls2.dfs.core.windows.net/bronze/"
```

```python
# first delta table! 
df_acct.write.format("delta").mode("overwrite").save(bronze_path + "accounts")
```

```python
# Checking the Bronze path
display(dbutils.fs.ls(bronze_path))
```

```python
display(dbutils.fs.ls(bronze_path + "accounts"))
```

```python
# Creating remaining delta tables 
df_cust.write.format("delta").mode("overwrite").save(bronze_path + "customers")
df_loan.write.format("delta").mode("overwrite").save(bronze_path + "loans")
df_tran.write.format("delta").mode("overwrite").save(bronze_path + "transactions")
df_ledg.write.format("delta").mode("overwrite").save(bronze_path + "ledger")

print("All Bronze Tables Created Successfully")
```
